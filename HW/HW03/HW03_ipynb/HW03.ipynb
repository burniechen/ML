```python
def UnivariateGaussian(mean, var):
    flag = True
    while flag:
        U = np.random.uniform(-1, 1)
        V = np.random.uniform(-1, 1)
        S = U**2 + V**2
        
        flag = False
        if S >= 1:
            flag = True

    coef = (-2 * np.log(S)) / S
    X = U * np.sqrt(coef)
    Y = V * np.sqrt(coef)
    
    return np.sqrt(var) * X + mean
```

```python
mean = 10
var = 2
points = []
for i in range(10000):
    points.append(UnivariateGaussian(mean, var))
```

```python
plt.figure(figsize=(10, 8))
plt.hist(points, 100)
```

```python
def PolynomialBasis(n, a, w):
    x = np.random.uniform(-1, 1)
    tmp = 0
    for i in range(n):
        tmp += w[i] * pow(x, i)
    e = UnivariateGaussian(0, a)
    y = tmp + e
    
    return y
```

```python
def GetXMatrix(x, n):
    x_m = []
    for i in range(n):
        x_m.append(pow(x, i))
    return np.array(x_m).reshape(1, 4)
```
